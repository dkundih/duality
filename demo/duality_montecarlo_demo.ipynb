```python
import duality #duality 3.3.7 was used in this example.
import pandas as pd #importing pandas to create a DataFrame.
```

```python
duality.__version__ #Version used.
```

```python
df = pd.DataFrame({
    'Example_data' : [38,37,38,36,36,35,36,34,33,34,35] #Example list of values.
})
```

```python
MC = duality.MonteCarlo(df['Example_data'], 50, 500) #Object defining the parameters of the simulation.
```

```python
MC.graph() #Creates a graph of the executed Monte Carlo simulation.
```

```python
MC.get_change() #Returns the list of percentage changes throughout the executed simulation.
```

```python
MC.get_risk(15000) #Shows the risk evaluation on top of the number of simulations defined in the parentheses.
```

```python
MC.get_stats() #Shows the statistics of the executed Monte Carlo simulation.
```

```python
MC.hist(graph_title='GitHub Repository Demo', x_title = 'Time period', y_title = 'Value') #Histogram of values.
```

```python
MC.hist(x_title = 'Time period', y_title = 'Value', method = 'e') #Histogram with Empirical rule applied, BEWARE: this option is currently being worked on so do not use it in any research!!!
```

```python
print(MC) #Explains the object.
```

```python
print(repr(MC)) #Explains the representation of the object.
```

```python
MC.get_logs() #Returns saved logs of executed functions. BEWARE: available only in DEVELOPER MODE.
```
